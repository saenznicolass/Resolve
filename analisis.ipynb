```python
df1 = pd.read_csv('Users.csv')
#df1.head()
```

```python
def no_utc(date):
    cad = date
    fecha = cad[:19]
    return fecha
df1['Date']=df1.apply(lambda x: no_utc(x['Date']), axis=1)
```

```python

df = pd.read_csv('Events.csv')
df['start_date'],df['end_date']=df.apply(lambda x: no_utc(x['start_date']), axis=1),df.apply(lambda x: no_utc(x['end_date']), axis=1)
```

```python
from time import time

###########Parámetro a cambiar#########
desface=1

def even(ini,end,med,name,event_df1):
    if (med>ini) and (med<end+datetime.timedelta(seconds=desface)):
        return name
    else:
        return event_df1
df1['event_name']='NN'
for i in range(257):
    start_time = time()
#    print(i)
    df1['event_name']=df1.apply(lambda x: even(dateutil.parser.isoparse(df['start_date'][i]),dateutil.parser.isoparse(df['end_date'][i]),dateutil.parser.isoparse(x['Date']),df['event_name'][i],x['event_name']), axis=1)
    end_time = time()
    
#    print('time=', end_time-start_time)
```
